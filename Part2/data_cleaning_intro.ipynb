Step1: Identify and Handle Missing Values 

1.Missing ages and replaces with average age of the users. 

2.Missing weight and height are replaced with the average of those values.

3.Missing values in 'purchases' are replaced with a number that we can decide later. 'None should be 0 assume it didn't buy'

```python
import json
import pandas as pd
import numpy as np
```

```python
#import data
df = pd.read_json('./data/data.json')
```

```python
df.head()
```

```python
# change name 'nan' to 'None'
df = df.replace({None: np.nan})
```

```python
df.head()
```

```python
#1.1 1.2 calculate average
average_age = df['age'].mean()
average_height = df['height_cm'].mean()
average_weight = df['weight_kg'].mean()
```

```python
print(average_age)
print(average_height)
print(average_weight)
```

```python
#1.1 1.2 fill in the missing value 
df['age']= df['age'].fillna(round(average_age)) 
df['height_cm']= df['height_cm'].fillna(round(average_height))
df['weight_kg']= df['weight_kg'].fillna(round(average_weight))
```

```python
df.head()
```

```python
#1.3 clean data 如果p是none. 就是0
def clean_purchases(purchases):
    return [0 if p is None else p for p in purchases]

#1.3 apply the 0 to purchases
df['purchases'] = df['purchases'].apply(clean_purchases)

df.head()
```

Step 2: Processing Outliers

```python
#setting a bonder/rule here
max_age = 100
max_weight = 150
```

```python
#replace any value that over max value
#using lamda

df['age'] = df['age'].apply(lambda x: min(x, max_age))
df['weight_kg'] = df['weight_kg'].apply(lambda x: min(x, max_weight))

df.head()
```

Step 3: Normallizaing the Data

```python
#start with maximum value 
# #using pandas
max_height = df['height_cm'].max()
max_weight = df['weight_kg'].max()
max_income = df['income'].max()

```

```python
#if the max value is 0, diveded by 0 will be a problem, so need change to 1
max_height = max_height if max_height !=0 else 1
max_weight = max_weight if max_weight !=0 else 1
max_income = max_income if max_income !=0 else 1
```

```python
# divide by maximum value 
df['height_cm'] = df['height_cm'] / max_height
df['weight_kg'] = df['weight_kg'] / max_weight
df['income'] = df['income'] / max_income
```

```python
# see the normalization data
df[['user','height_cm','weight_kg', 'income']].head()
```
